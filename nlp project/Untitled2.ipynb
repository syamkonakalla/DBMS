```python
myfile = open("data.txt", "r")
```

```python
myfile.seek(0)
mylist = myfile.readlines()
print(mylist)
```

```python
string=''
myli=[]
for i in mylist:
    if i=='\n':
        myli.append(string)
        string=''
        continue
    else:
        string = string + i
print(myli)
```
